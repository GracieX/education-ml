```python
import pandas as pd
```

# Column Name Descriptions  
INSTNM (Institution Name)  
SATBBR (State Abbreviation)  
City  
Latitude  
Longitude  
SCHTYPE (1=Public, 2=Private,non-profit)  
C100_4 (completion rate for 4 year institutions)  
TRANS_4 (Transfer rate for first-time, full-time students at four-year institutions)  
SAT_AVG (SAT average score)  
ACTCMMID (ACT composite midpoint score)  
UGDS_WHITE (Total share of enrollment of undergraduate degree-seeking students who are white)  
UGDS_BLACK (Total share of enrollment of undergraduate degree-seeking students who are Black)  
UGDS_HISP (Total share of enrollment of undergraduate degree-seeking students who are Hispanic)  
UGDS_ASIAN (Total share of enrollment of undergraduate degree-seeking students who are Asian)

```python
SY19 = pd.read_csv("MERGED2018_19_PP.csv")
# SY19.IL()
SY19_IL = SY19.loc[SY19["STABBR"]=="IL"]
# SY19_IL.head()
SY19_IL = SY19_IL[["INSTNM","STABBR","CITY","LATITUDE","LONGITUDE","SCHTYPE","C100_4","TRANS_4","SAT_AVG","ACTCMMID","UGDS_WHITE","UGDS_BLACK","UGDS_HISP","UGDS_ASIAN","UGDS_AIAN"]]
SY19_IL["Year"] = "2019"
# SY19_IL.head()
SY19_IL = SY19_IL.dropna()
SY19_IL.head()
```

```python
SY18 = pd.read_csv("MERGED2017_18_PP.csv")
# SY18.head()
SY18_IL = SY18.loc[SY18["STABBR"]=="IL"]
# SY18_IL.head()
SY18_IL = SY18_IL[["INSTNM","STABBR","CITY","LATITUDE","LONGITUDE","SCHTYPE","C100_4","TRANS_4","SAT_AVG","ACTCMMID","UGDS_WHITE","UGDS_BLACK","UGDS_HISP","UGDS_ASIAN","UGDS_AIAN"]]
SY18_IL["Year"]="2018"
# SY18_IL.head()
```

```python
for index, row in SY18_IL.iterrows():
    if row['INSTNM'] in SY19_IL['INSTNM'].tolist():
        latitude = SY19_IL.loc[SY19_IL['INSTNM']==row['INSTNM'], 'LATITUDE'].iloc[0]
        SY18_IL.at[index,'LATITUDE']= latitude
        longitude = SY19_IL.loc[SY19_IL['INSTNM']==row['INSTNM'], 'LONGITUDE'].iloc[0]
        SY18_IL.at[index,'LONGITUDE']= longitude
        school_type = SY19_IL.loc[SY19_IL['INSTNM']==row['INSTNM'], 'SCHTYPE'].iloc[0]
        SY18_IL.at[index,'SCHTYPE']= school_type
        
SY18_IL = SY18_IL.dropna()
SY18_IL.head()
```
